## Signal Frontier Analysis

"Mean reversion is the assumption that a stock's price will tend to move to the average price over time." [1] With this strategy, we will review a list of stocks for their cummulative returns and then compute a portfolio for a particular lookback in time and days between trading (holding period) to determine the Sharpe ratio. [2] "The Sharpe Ratio is a measure for calculating risk-adjusted return and is considered the industry standard for such calculations." [3] It is defined as the mean portfolio return minus risk-free rate divided by standard deviation of portfolio return; i.e., Sharpe Ration = $\frac{\bar{r_p}-r_f}{\rho_p}$ where: 
- $\bar{r_p}$ = Expected portfolio return.
- $r_f$ = Risk free rate
- $\rho_p$ = Portfolio standard deviation

## Retrieve stock data for a list of stock symbols

```python
def stock_returns(panel_data, freq):
    """
    convert stock 'Close' data dataframe timeseries to specified frequency with 'ffill' method
    """
    # 'B' = business, 'ffill' = propagate last valid observation forward to next valid
    stock_data = panel_data['Close'].asfreq(freq, method='ffill')

    # returns
    return stock_data.pct_change()
```

## Calculate Sharpe Ratio for Stock Portfolio

```python
def calc_mom(price, lookback, lag=1):
    """
    A tranform function to compute mean reversion over a certain lookback,
        then rank in descending order and standardize.
    """
    # shift index by lag periods and determine percent change
    mom_ret = price.shift(lag).pct_change(lookback)
    # compute numerical data ranks along column in descending order
    ranks = mom_ret.rank(axis=1, ascending=False)
    # subtract off means 
    demeaned = ranks.subtract(ranks.mean(axis=1), axis=0)
    # divide by standard deviation
    return demeaned.divide(demeaned.std(axis=1), axis=0)
```

```python
compound = lambda x: (1 + x).prod() - 1
daily_sr = lambda x: x.mean() / x.std()

def strat_sr(prices, lb, hold):
    """
    A strategy to return a Sharpe Ratio given stock prices, loopback and holding period.
    """
    # compute protfolio weights; e.g., 30 for hold will create a freq = '30B' period
    freq = '%dB' % hold
    # create new portfolio
    port = calc_mom(prices, lb)
    # daily returns
    daily_rets = prices.pct_change()
    
    # compute portfolio returns
    port = port.shift(1).resample(freq).first()
    returns = daily_rets.resample(freq).apply(compound)
    port_rets = (port * returns).sum(axis=1)
    
    # 252 = busines days
    return daily_sr(port_rets) * np.sqrt(252 / hold)
```

```python
strat_sr(stocks['Close'], 70, 30)
```

## Visualization of Sharpe Ratio

```python
def heatmap(df, cmap=plt.cm.gray_r):
    """
    A function to visualize results of Sharpe Ratio for parameters grid
    """
    fig = plt.figure()
    ax = fig.add_subplot(111)
    # imshow - plot numpy array as images
    axim = ax.imshow(df.values, cmap=cmap, interpolation='nearest')
    ax.set_xlabel(df.columns.name)
    ax.set_xticks(np.arange(len(df.columns)))
    ax.set_xticklabels(list(df.columns))
    ax.set_ylabel(df.index.name)
    ax.set_yticks(np.arange(len(df.index)))
    ax.set_yticklabels(list(df.index))
    plt.colorbar(axim)
```

```python
# evaluate strategy Sharpe Ratio function of a grid of parameters
from collections import defaultdict

lookbacks = range(20, 90, 5)
holdings = range(20, 90, 5)
dd = defaultdict(dict)
for lb in lookbacks:
    for hold in holdings:
        dd[lb][hold] = strat_sr(stocks['Close'], lb, hold)
        
ddf = DataFrame(dd)
ddf.index.name = 'Holding Period'
ddf.columns.name = 'Lookback Period'

heatmap(ddf)
```

Heatmap of mean reversion strategy Sharpe Ratio (higher is better) over various lookbacks and holding periods.
